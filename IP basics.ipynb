```python
import cv2
import glob
```

```python
path="/home/shikhar/Desktop/git_workspace/ImageProcessing_Practice-/image folder/*"
```

```python
for file in glob.glob(path):
    print(file)
    a=cv2.imread(file)
    print(a)
    c=cv2.cvtColor(a,cv2.COLOR_BGR2RGB)
    cv2.imshow("images",c)
    
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
