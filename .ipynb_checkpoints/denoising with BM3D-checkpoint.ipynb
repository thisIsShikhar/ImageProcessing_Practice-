```python
from skimage import io, img_as_ubyte
from matplotlib import pyplot as plt
import bm3d
```

```python
img =io.imread("./oct.jpg", as_gray=True)
```

```python
denoised= bm3d.bm3d(img+0.5, sigma_psd=30/255, stage_arg=bm3d.BM3DStages.HARD_THRESHOLDING)
```

```python
plt.imshow(denoised)
# plt.imsave("/home/shikhar/Desktop/git_workspace/ImageProcessing_Practice-/denoising/bm3d.jpg", denoised)
```
