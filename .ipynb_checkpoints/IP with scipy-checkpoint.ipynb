```python
from skimage import io, img_as_ubyte
from scipy import ndimage
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = img_as_ubyte(io.imread("/home/shikhar/Desktop/git_workspace/ImageProcessing_Practice-/random-pixels-big.png", as_gray=False))
```

```python
print(img.size, img.shape, img.dtype)
```

```python
max_value=img.max()
min_value=img.min()
mean_value=img.mean()
```

```python
print(max_value, min_value,mean_value)
```

```python
flippedLR=np.fliplr(img)
flippedUD=np.flipud(img)
```

```python
plt.subplot(2,1,1)
plt.imshow(img, cmap="Greys")

plt.subplot(2,2,3)
plt.imshow(flippedLR, cmap="Blues")

plt.subplot(2,2,4)
plt.imshow(flippedUD, cmap='hsv')
```

```python
rotated=ndimage.rotate(img,45,reshape=False)
plt.imshow(rotated)
```

use of filters

```python
uniform_filtered=ndimage.uniform_filter(img,size=15)  #used for bluring images
plt.imshow(uniform_filtered)
```

```python
gaussian_filtered=ndimage.gaussian_filter(img,sigma=5) #used for bluring or smoothing. It gradually removes the edges.
plt.imshow(gaussian_filtered)
```

```python
median_filtered=ndimage.median_filter(img,5)  #used for denoising images
plt.imshow(median_filtered)
```

```python
sobel_img= ndimage.sobel(img) #used for detecting edges
plt.imshow(sobel_img)
```
