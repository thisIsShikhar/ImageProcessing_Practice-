```python
from skimage.filters import scharr,prewitt,sobel, roberts


img=io.imread("/home/shikhar/Desktop/git_workspace/ImageProcessing_Practice-/test_image.jpg", as_gray=True)


edge_roberts=roberts(img)
edge_sobel=sobel(img)
edge_scharr=scharr(img)
edge_prewitt=prewitt(img)
```

Deconvulation

```python
from skimage import restoration 
import numpy as np
import scipy.stats as st
```

```python
image=io.imread("/home/shikhar/Desktop/git_workspace/ImageProcessing_Practice-/processed images/cropped with pillow.png", as_gray=True)
```

```python
print(image.shape)
```

```python
def gkern(kernlen=21, nsig=3):
    """Returns a 2D Gaussian kernel."""

    lim = kernlen//2 + (kernlen % 2)/2
    x = np.linspace(-lim, lim, kernlen+1)
    kern1d = np.diff(st.norm.cdf(x))
    kern2d = np.outer(kern1d, kern1d)
    return kern2d/kern2d.sum()

psf=gkern(5,4)
print(psf)
deconvoled= restoration.unsupervised_wiener(image, psf)
```

entropy images (texture based)

```python
import matplotlib.pyplot as plt
from skimage import io, restoration
from skimage.filters.rank import entropy
from skimage.morphology import disk
```

```python
img=io.imread("/home/shikhar/Desktop/git_workspace/ImageProcessing_Practice-/scratch.jpg", as_gray=True)
```

```python
entropy_image=entropy(img, disk(3))
```

```python
plt.imshow(entropy_image, cmap="gray")
```

Threshold

```python
from skimage.filters import try_all_threshold
```

```python
fig, ax =try_all_threshold(entropy_image, figsize=(10,8), verbose= False)
plt.show()
```

```python
from skimage.filters import threshold_otsu
import numpy as np 
```

```python
otsu=threshold_otsu(entropy_image)

binary=entropy_image <= otsu
```

```python
plt.imshow(binary, cmap="gray")
```

```python
print("The percentage of bright pixels: ",(np.sum(binary==1)*100/(np.sum(binary==1)+np.sum(binary==0))))
```
