Gaussian filter

```python
from skimage import io, img_as_float
from scipy import ndimage as nd
from matplotlib import pyplot as plt
import numpy as np
```

```python
img=img_as_float(io.imread("oct.jpg"))
```

```python
gaussian=nd.gaussian_filter(img, sigma=4.2)
plt.imshow(gaussian,cmap='gray')
```

median filter

```python
median=nd.median_filter(img, size=13)
plt.imshow(median,cmap='gray')
```

non local means filter

```python
from skimage.restoration import denoise_nl_means, estimate_sigma
```

```python
sigma_est= np.mean(estimate_sigma(img, multichannel=True))
denoise= denoise_nl_means(img,h=0.5*sigma_est, fast_mode=True, patch_size=3, patch_distance=2, multichannel=True)
```

```python
plt.imshow(denoise)
```
